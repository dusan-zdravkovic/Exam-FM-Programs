- Scenario: You borrow $10,000 for 5 years at an annual interest rate of 5%.
You will repay the loan with equal payments at the end of each year (Annuity-Immediate).
We want to calculate:
	1.	The annual payment amount.
	2.	The outstanding loan balance after each payment.
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
loan_amount = 10000  # Initial loan amount
interest_rate = 0.05  # Annual interest rate (5%)
num_years = 5  # Duration of loan in years

# Calculate the payment using the annuity-immediate formula
def calculate_annuity_payment(loan, rate, years):
    return loan * rate / (1 - (1 + rate) ** -years)

# Annual payment
payment = calculate_annuity_payment(loan_amount, interest_rate, num_years)
print(f"Annual Payment: ${payment:.2f}")

# Initialize lists to store data for plotting
balances = [loan_amount]
interest_paid = []
principal_paid = []

# Calculate the breakdown of each payment
outstanding_balance = loan_amount
for year in range(1, num_years + 1):
    interest = outstanding_balance * interest_rate
    principal = payment - interest
    outstanding_balance -= principal
    
    # Store values for plotting
    balances.append(outstanding_balance)
    interest_paid.append(interest)
    principal_paid.append(principal)

# Plot the outstanding balance over time
plt.figure(figsize=(12, 6))
plt.plot(range(num_years + 1), balances, marker='o', linestyle='--', label="Outstanding Balance")
plt.bar(range(1, num_years + 1), interest_paid, width=0.4, label="Interest Paid", alpha=0.6, color='orange')
plt.bar(range(1, num_years + 1), principal_paid, width=0.4, bottom=interest_paid, label="Principal Paid", alpha=0.6, color='green')
plt.title("Loan Outstanding Balance with Interest and Principal Breakdown")
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.xticks(range(num_years + 1))
plt.grid(True)
plt.legend()
plt.show()
```

You want to save $20,000 for a car in 5 years. Youll contribute monthly at the beginning of each month to an account earning 6% annual interest compounded monthly. How much should you contribute each month?

```python
# Parameters
future_value_goal = 20000  # Goal: $20,000
annual_interest_rate = 0.06  # 6% annual interest
num_years = 5  # Timeframe: 5 years
num_payments = num_years * 12  # Monthly payments
monthly_interest_rate = annual_interest_rate / 12  # Monthly interest rate

# Future Value of an Annuity-Due Formula: FV = P * ((1 + i)^n - 1) / i * (1 + i)
def calculate_monthly_payment(fv_goal, rate, periods):
    return fv_goal / (((1 + rate) ** periods - 1) / rate * (1 + rate))

# Calculate the required monthly payment
monthly_payment = calculate_monthly_payment(future_value_goal, monthly_interest_rate, num_payments)
print(f"Required Monthly Payment: ${monthly_payment:.2f}")

# Plot how the savings grow over time
balances = [0]
for month in range(1, num_payments + 1):
    balances.append(balances[-1] * (1 + monthly_interest_rate) + monthly_payment)

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(range(num_payments + 1), balances, marker='o', linestyle='--', label="Savings Balance")
plt.axhline(y=future_value_goal, color='r', linestyle='-', label="Goal ($20,000)")
plt.title("Growth of Savings Over Time")
plt.xlabel("Months")
plt.ylabel("Savings Balance ($)")
plt.xticks(range(0, num_payments + 1, 12))
plt.grid(True)
plt.legend()
plt.show()
```
