```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

### Time Value of Money Problem

```python
# Parameters
principal_x = 100  # Initial deposit for Fund X
principal_y = 200  # Initial deposit for Fund Y
i_eff_x = 0.12749  # Effective annual interest rate for Fund X
delta_y = 0.08     # Annual force of interest for Fund Y
time = np.linspace(0, 20, 100)  # Time range (0 to 20 years)

# Fund X balance (compounded annually)
balance_x = principal_x * (1 + i_eff_x) ** time

# Fund Y balance (continuous compounding)
balance_y = principal_y * np.exp(delta_y * time)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, balance_x, label="Fund X (Nominal Discount)", color="blue")
plt.plot(time, balance_y, label="Fund Y (Force of Interest)", color="green")
plt.axvline(17.3, color="red", linestyle="--", label="n ≈ 17.3 years")
plt.xlabel("Time (years)")
plt.ylabel("Balance")
plt.title("Balance Growth of Fund X and Fund Y")
plt.legend()
plt.grid()
plt.show()
```
