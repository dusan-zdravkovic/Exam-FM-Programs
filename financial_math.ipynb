```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

### Time Value of Money Problem

```python
# Parameters
principal_x = 100  # Initial deposit for Fund X
principal_y = 200  # Initial deposit for Fund Y
i_eff_x = 0.12749  # Effective annual interest rate for Fund X
delta_y = 0.08     # Annual force of interest for Fund Y
time = np.linspace(0, 20, 100)  # Time range (0 to 20 years)

# Fund X balance (compounded annually)
balance_x = principal_x * (1 + i_eff_x) ** time

# Fund Y balance (continuous compounding)
balance_y = principal_y * np.exp(delta_y * time)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, balance_x, label="Fund X (Nominal Discount)", color="blue")
plt.plot(time, balance_y, label="Fund Y (Force of Interest)", color="green")
plt.axvline(17.3, color="red", linestyle="--", label="n ≈ 17.3 years")
plt.xlabel("Time (years)")
plt.ylabel("Balance")
plt.title("Balance Growth of Fund X and Fund Y")
plt.legend()
plt.grid()
plt.show()
```

```python
# Annuities
# Parameters
payment = 100  # Payment amount
n_periods = 10  # Number of payments
interest_rate = 0.05  # Annual interest rate

# Calculate Present Value for Annuity-Immediate
v = 1 / (1 + interest_rate)  # Discount factor
PV_annuity_immediate = payment * (1 - v**n_periods) / interest_rate

# Calculate Present Value for Annuity-Due
PV_annuity_due = PV_annuity_immediate * (1 + interest_rate)

# Plot the two annuities
time = range(1, n_periods + 1)
payments = [payment] * n_periods

plt.figure(figsize=(10, 6))

# Annuity-Immediate: Payments at the end of each period
plt.plot(time, payments, label="Annuity-Immediate (Payments End of Period)", linestyle="--", marker="o")

# Annuity-Due: Payments at the start of each period
plt.plot([t - 1 for t in time], payments, label="Annuity-Due (Payments Start of Period)", linestyle="-", marker="s")

# Labels and legend
plt.title("Comparison of Annuity-Immediate vs. Annuity-Due")
plt.xlabel("Time Period")
plt.ylabel("Payment Amount")
plt.xticks(range(n_periods + 1))  # Include time 0 for Annuity-Due
plt.legend()
plt.grid()
plt.show()

# Print results
print(f"Present Value of Annuity-Immediate: ${PV_annuity_immediate:.2f}")
print(f"Present Value of Annuity-Due: ${PV_annuity_due:.2f}")
```

```python
# Parameters
payment = 100  # Payment amount
n_periods = 10  # Number of payments
interest_rate = 0.05  # Annual interest rate

# Time periods
time_annuity_immediate = range(1, n_periods + 1)  # Payments start at t=1
time_annuity_due = range(0, n_periods)  # Payments start at t=0

# Cumulative payment values over time
cumulative_immediate = [payment * (1 - (1 / (1 + interest_rate))**t) / interest_rate for t in time_annuity_immediate]
cumulative_due = [payment * (1 - (1 / (1 + interest_rate))**(t + 1)) / interest_rate for t in time_annuity_due]

# Plotting
plt.figure(figsize=(12, 6))

# Cumulative Value for Annuity-Immediate
plt.plot(time_annuity_immediate, cumulative_immediate, label="Cumulative Value: Annuity-Immediate", marker="o", linestyle="--")

# Cumulative Value for Annuity-Due
plt.plot(time_annuity_due, cumulative_due, label="Cumulative Value: Annuity-Due", marker="s", linestyle="-")

# Highlight timing difference
plt.axvline(x=0, color="gray", linestyle=":", label="Annuity-Due Starts at t=0")
plt.axvline(x=1, color="black", linestyle=":", label="Annuity-Immediate Starts at t=1")

# Labels and customization
plt.title("Cumulative Present Value Comparison: Annuity-Immediate vs. Annuity-Due")
plt.xlabel("Time Period")
plt.ylabel("Cumulative Present Value ($)")
plt.xticks(range(n_periods + 1))  # Include time 0
plt.legend()
plt.grid()
plt.show()

# Print final present values
PV_annuity_immediate = cumulative_immediate[-1]
PV_annuity_due = cumulative_due[-1]
print(f"Final Present Value (Annuity-Immediate): ${PV_annuity_immediate:.2f}")
print(f"Final Present Value (Annuity-Due): ${PV_annuity_due:.2f}")
```
